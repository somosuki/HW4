```python
# Dependencies
import pandas as pd
import os
```

```python
# Make a reference to the csv file path
schools_path = os.path.join("raw_data", "schools_complete.csv")
# Import the csv file as a DataFrame
schools_df = pd.read_csv(schools_path, encoding="utf-8")
schools_df.head()
```

```python
renamed_df = schools_df.rename(columns={"name":"school"})
renamed_df.head()
```

```python
students_path = os.path.join("raw_data", "students_complete.csv")
students_df = pd.read_csv(students_path, encoding="utf-8")
students_df.head()
```

```python
# Merge two dataframes using an inner join
merge_table = pd.merge(renamed_df, students_df, on="school")
merge_table.head()
```

```python
# Group merge_table by the "type" column
type_df = merge_table.groupby("school")
type_df.count()
```
